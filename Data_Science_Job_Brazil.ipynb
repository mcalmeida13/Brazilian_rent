```
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import json
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory
```

```
filename = "/content/drive/My Drive/Data Science/Artigos/Data-Science-Brazil/glassdoor_jobs.csv"
df = pd.read_csv(filename)
```

```
df.head()
```

```
#Per hour columns
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```
#Employee provided
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
```

```
df = df[df['Salary Estimate'] != '-1']
```

```
#get rid of Glassdoor est.
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```
# removing K and $
minus_kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```
#removing remaining per hour
min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:',''))
```

```
#min salary
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['min_salary'].dtype
```

```
#max salary
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
df['max_salary'].dtype
```

```
df['avg_salary'] = (df.min_salary + df.max_salary)/2
```

```
df.head()
```

```
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-5], axis =1 )
```

```
df.head()
```

```
# df['city'] = df['Location'].apply(lambda x: x.split(',')[0])
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```
#jobs by states
df.job_state.value_counts()
```

```
df.head()
```

```
df['job_head'] = df['Headquarters'].apply(lambda x: x.split(',')[-1])
```

```
df.head()
```

```
df['same_state'] = df.apply(lambda x: 1 if x.job_state == x.job_head else 0, axis =1)
```

```
df.head()
```

```
#age of the company

df['age'] = df.Founded.apply(lambda x: x if x < 1 else 2020 - x)
```
